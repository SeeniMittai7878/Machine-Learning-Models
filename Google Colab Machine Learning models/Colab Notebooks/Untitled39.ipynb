```python
#feature plots in on target


data.columns
```

```python
interset = ['acousticness', 'danceability', 'duration_ms', 'energy',
       'instrumentalness', 'key', 'liveness', 'loudness', 'mode',
       'speechiness', 'tempo', 'time_signature', 'valence', 'target',]
```

```python
for name in interset:
  pos_data = data[data['target'] == 1][name]
  nav_data = data[data['target'] == 0][name] 

  plt.figure(figsize=(12,7))

  sns.histplot(pos_data,bins = 30,label = "positive",color = "green")
  sns.histplot(nav_data,bins = 30,label = "negative",color = "red")
  plt.legend()
  plt.title(f'positive negative histogramm plots for {name}')
```

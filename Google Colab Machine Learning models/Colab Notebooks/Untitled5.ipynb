```python
import seaborn 
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

x = np.array([1,2,3,4,5])
y= np.array([20,40,60,80,100])
plt.plot(x,y,marker = "*",)
plt.show()


years = np.array([2011,2012,2013,2014,2015,2016])
yeild_apples = np.array(["20%","60%","50%","42%","72%","80%"])
plt.plot(years,yeild_apples)
plt.xlabel('year')
plt.ylabel('prsantage')
plt.show()




year = range(2000,2008)
apples = [0.895,0.915,0.920,0.935,0.945,0.953,0.965,0.975]
orange = [0.23,0.927,0.933,0.937,0.942,0.948,0.948,0.955]

plt.figure(figsize=(8,6))
plt.plot(year,apples)
plt.plot(year,orange)

plt.title('corp yeild on stage')
plt.legend('apples','orages')







#rgdgjhf
plt.scatter(year,apples)
plt.scatter(year,orange)
```
